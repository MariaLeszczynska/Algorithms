# Task 1


```python
'''Create a function which for a given list of integers verifies whether it contains a sequence of k consecutive numbers.'''
def has_consecutive_sequence(lst, k):
    lst.sort()
    consecutive_count = 1
    for i in range(1, len(lst)):
        if lst[i] == lst[i - 1] + 1:
            consecutive_count += 1
            if consecutive_count >= k:
                return True
        else:
            consecutive_count = 1

    return False

my_list = [1, 2, 3, 5, 6, 7, 9]
k = 4
result = has_consecutive_sequence(my_list, k)
if result:
    print("The list contains a sequence of {k} consecutive numbers.")
else:
    print("The list does not contain a sequence of {k} consecutive numbers.")
```

```python
def remove_polish_diacritics(list_of_strings):
    new_list = []
    polish_diacritics = ['ą', 'ę' 'ź', 'ć', 'ł', 'ż', 'ś', 'ó', 'ń']
    for i in list_of_strings:
        remove = False
        for char in i:
            if char in polish_diacritics:
                remove = True
        if not remove:
            new_list.append(i)
    return new_list

names = ['Kaczyński', 'Tusk', 'Hołownia', 'Kosiniak-Kamysz', 'Czarzasty', 'Mentzen']
remove_polish_diacritics(names)
```

```python
'''For a given list of strings consisting of only small and capital letters of the latin alphabet create a function which for each
string generates its version such that each capital letter in this string is preceded by a whitespace.'''
def add_whitespaces(without_whitespaces):
    with_whitespaces = [concatenate]
    for sentence in without_whitespaces:
         s = sentence[0]
         for j in range(1, len(sentence)):
             if sentence[j].isupper():
                 s += ' ' + sentence[j]
             else:
                 s += sentence[j]
         with_whitespaces.append(s)
    return with_whitespaces
without_whitespaces = ['jnidsiNJbBBTNjnnI']
add_whitespaces(without_whitespaces)
```

# Computational complexity

```python
import random
import copy
import timeit
import matplotlib.pyplot as plt


def three_greatest_elements(lst):
    max_values = [float('-inf')] * 3
    for num in lst:
        if num > max_values[0]:
            max_values = [num, max_values[0], max_values[1]]
        elif num > max_values[1]:
            max_values = [max_values[0], num, max_values[1]]
        elif num > max_values[2]:
            max_values = [max_values[0], max_values[1], num]
    return max_values

def experiment(list_length):
    random.seed(42)
    aList = [random.randint(1, 1000000) for _ in range(list_length)]
    bList = copy.deepcopy(aList)

    aList_sort_time = timeit.timeit(lambda: sorted(aList)[-3], number=10)

    bList_custom_time = timeit.timeit(lambda: three_greatest_elements(bList), number=10)

    return aList_sort_time, bList_custom_time


list_lengths = [10000, 50000, 100000, 200000, 300000, 400000, 500000]

aList_sort_times = []
bList_custom_times = []

for length in list_lengths:
    aList_sort_time, bList_custom_time = experiment(length)
    aList_sort_times.append(aList_sort_time)
    bList_custom_times.append(bList_custom_time)

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(list_lengths, aList_sort_times, label="Sort (aList)")
plt.plot(list_lengths, bList_custom_times, label="Custom (bList)")
plt.xlabel("List Length")
plt.ylabel("Time (seconds)")
plt.legend()
plt.title("Execution Time vs. List Length")
plt.show()
```
